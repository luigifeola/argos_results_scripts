# Change date of all subfolders giving dates and folder path

```python
parent_dir = '/home/luigi/Desktop/test'
wrong_date = '2020-07-25'
right_date = '2020-07-01'
```

```python
from datetime import datetime
import os
import os.path
import sys
dir_path = os.path.abspath(parent_dir)
print(dir_path)
print(os.listdir(dir_path))
for dirName, subdirList, fileList in os.walk(parent_dir):
    for folder_name in subdirList:
        old_path_name = os.path.join(dirName, folder_name)
        elements = folder_name.split("_")
        if elements[0] == wrong_date:
            elements[0] = right_date
        else:
            continue
        new_folder_name = '_'.join(elements)
        new_path_name = os.path.join(dirName, new_folder_name)
        os.rename(old_path_name, new_path_name)
```
