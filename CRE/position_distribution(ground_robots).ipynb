```python
import os
import pandas as pd
from termcolor import colored
import numpy as np
from collections import OrderedDict
import matplotlib.pyplot as plt
from natsort import natsorted
import sys
import pickle
import seaborn as sns

# Display pandas df without truncation
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)

np.set_printoptions(edgeitems=30, linewidth=100000, formatter=dict(float=lambda x: "%.3g" % x))

```

```python
fieldsPerKilobot = 4
num_robot = 80
```

```python
bin_edges = np.linspace(-1, 1, 21)
bin_edges
```

```python
# results_folder = '/home/luigi/Documents/ARGoS/argos_results_scripts/CRE/results_cre/different_resources_number/different_resources_2021-05-26_900seconds/'
results_folder = '/home/luigi/Documents/ARGoS/argos_results_scripts/CRE/results_cre/equal_resources_number/equal_resources_2021-05-27_seconds#900/'
fileName = "seed#1_areaLOG_client.tsv"
```

```python
if not os.path.isdir(results_folder):
    print(colored("Error, " + results_folder + " does not exist", 'red'))
else:
    print(colored("OK, " + results_folder + " exists", 'green'))
```

```python
kilo_pos_label = []
for i in range(num_robot):
    kilo_pos_label += ["posy"+str(i)]
# area_pos_label

print(kilo_pos_label)
```

```python
df_full_position = pd.DataFrame()
print(colored("Main folder:", 'blue'),results_folder)
for filename in natsorted(os.listdir(results_folder)):
#     print('\t filename: ', filename)
    seed = filename.split("_")[0]
#     if seed == 'seed#3':
#         print(seed.split("#")[-1])
#         break

    if filename.endswith("kiloLOG_client.tsv"):
        if not os.path.getsize(os.path.join(results_folder, filename)) > 0:
            print(colored("\tWARNING, empty file at:" + filename, 'red'))
            continue
        print('\tfilename: ', filename)
        df_kilo_client = pd.read_csv(os.path.join(results_folder, filename), sep="\t",
                                     header=None)
    
        if(len(df_kilo_client.columns)==322):
            print("\tCutting null elements in client kilo df")
            df_kilo_client.drop(df_kilo_client.columns[len(df_kilo_client.columns) - 1], axis=1, inplace=True)
            
        col_kilo_labels = ['time']
        for i in range(0, len(df_kilo_client.columns) - 1, fieldsPerKilobot):
            col_kilo_labels += ['id' + str(i // fieldsPerKilobot), 'posx' + str(i // fieldsPerKilobot), 'posy' + str(i // fieldsPerKilobot), 'ori' + str(i // fieldsPerKilobot)]

        df_kilo_client.columns = col_kilo_labels
    
        if df_full_position.empty:
            df_full_position = df_kilo_client[kilo_pos_label]
        else:
            df_full_position = df_full_position.join(df_kilo_client[kilo_pos_label], lsuffix='', rsuffix='_'+seed.split("#")[-1])
            
            
        hist_val = np.array([])
        for x in df_kilo_client[kilo_pos_label].values:
        #     print(x.size)
            hist,_= np.histogram(x,bin_edges)
            hist_val = np.vstack([hist_val, hist]) if hist_val.size else hist
            
            
        values = np.flip(hist_val.T, 0)
        y_edges = np.flip(bin_edges.round(decimals=2))

        fig = plt.figure(figsize=(10, 5), dpi=80)
        # plt.xlabel('time(s)')
        yticks = y_edges



        # print("\t v_max:", v_max)
        ax = sns.heatmap(values, yticklabels=yticks, cmap="viridis", vmin=0, vmax=40)
        ax.set_title("Position distribution")
        ax.set_xlabel('time(s)')
        plt.ylabel('y-position')
        plt.yticks(np.arange(y_edges.size), yticks)
        plt.savefig("position_distribution_equal_resources"+seed+".png")
        plt.show()
```
