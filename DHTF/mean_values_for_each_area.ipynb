```python
experiment = "random_mixed"
```

```python
results_folder = os.path.join(os.getcwd(), "results_adaptive/"+experiment)
```

```python
if not os.path.isdir(results_folder):
    print(colored("Error, results_folder does not exist", 'red'))
else:
    print(colored("OK, "+results_folder+" exists", 'green'))
```

```python
for timeout_folder in natsorted(os.listdir(os.path.join(results_folder))):
    print(colored("Timeout folder:", 'blue'),timeout_folder)
    parameters = timeout_folder.split("_")
    
    for param in parameters:
        if param.startswith("timeout"):
            timeout = int(param.split("#")[-1]) * 10
            # print("\ttimeoutR:",timeoutR)

    if timeout == -1 :
        print(colored("\tWARNING: wrong timeout folder", 'green'))
        continue
        
    for filename in os.listdir(os.path.join(results_folder, timeout_folder)):
        if filename.endswith("areaLOG_server.tsv"):
            if not os.path.getsize(os.path.join(results_folder, timeout_folder, filename)) > 0:
                print(colored("\tWARNING, empty file at:" + filename, 'red'))
                continue
            print('\tfilename: ', filename)
            
            df_server = pd.read_csv(os.path.join(results_folder, timeout_folder, filename), sep="\t", header=None)
       
        if filename.endswith("areaLOG_client.tsv"):
            if not os.path.getsize(os.path.join(results_folder, timeout_folder, filename)) > 0:
                print(colored("\tWARNING, empty file at:" + filename, 'red'))
                continue
            print('\tfilename: ', filename)
            
            df_client = pd.read_csv(os.path.join(results_folder, timeout_folder, filename), sep="\t", header=None)
            
            break
    break
```

#### Generate col labels

```python
col_labels = ['time']
for i in range (0, len(df_server.columns)-2, 6):
    print(i,end=", ")
    col_labels += ['id'+str(i//6), 'posx'+str(i//6), 'posy'+str(i//6), 'color'+str(i//6), 'completed'+str(i//6), 'contained'+str(i//6)]
```

#### Remove last empty col and assigni labels to df_server

```python
if len(df_server.columns) > 49:
    print("Cutting null elements")
    df_server.drop(df_server.columns[len(df_server.columns)-1], axis=1, inplace=True)
df_server.columns = col_labels
```

#### First df_client row contains garbage, so is substituited with the second row except for the time, then remove Nan valuse in [:,49:]

```python
if len(df_client.columns) > 49:
    print("Cutting null elements")
    df_client.loc[0,1:] = df_client.loc[1,1:] 
    df_client = df_client.drop(np.arange(49,len(df_client.columns)), axis=1)
```

#### Assign the same label of df_server to df_client

```python
df_client.columns = col_labels
```
